```python
sales_path = 'dbt_banxware_assignment/seeds/raw_sales_data.csv'
customers_path = 'dbt_banxware_assignment/seeds/raw_customer_data.csv'
```

```python
df_sales = pd.read_csv(sales_path)
df_customers = pd.read_csv(customers_path)
```

```python
df_sales['order_date'] = pd.to_datetime(df_sales['order_date'], format='%m/%d/%Y')
df_clean = df_sales[df_sales['order_status'] == 'Completed'].copy()
df_clean['total_sales_amount'] = df_clean['quantity'] * df_clean['price']
df_clean['year'] = df_clean['order_date'].dt.year
df_clean['month'] = df_clean['order_date'].dt.month
df_2023 = df_clean[df_clean['year'] == 2023].copy()
df_2023.info()
```

```python
# Top 5 Products
q1 = df_2023.groupby('product_id')['total_sales_amount'].sum().sort_values(ascending=False).head(5)
q1
```

```python
# Top 5 Customers
df_merged = df_2023.merge(df_customers, left_on='customer_id', right_on='id')
q2 = df_merged.groupby('name')['total_sales_amount'].sum().sort_values(ascending=False).head(5)
q2
```

```python
# Avg Order Value per Month
q3 = df_2023.groupby('month')['total_sales_amount'].mean().sort_index()
q3
```

```python
# Highest Volume Customer
df_oct = df_merged[df_merged['month'] == 10]
q4 = df_oct['name'].value_counts().head(1)
q4
```
